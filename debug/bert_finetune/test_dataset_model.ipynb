```python
import sys
sys.path.append("../../src/")
```

```python
from transformers import AutoTokenizer
```

```python
from llm2bert.bert_finetune.arguments import DataArguments, ModelArguments
from llm2bert.bert_finetune.data import ClassificationDataset, DiyDataCollator
from llm2bert.bert_finetune.model import SequenceClassificationModel
```

```python
diy_dataset = ClassificationDataset(
    csv_file=r"../../examples/multi_classification/llm_parsed_results.csv",
    bert_format="给下述文本分类：${text}",
)
```

```python
diy_dataset[0]
```

```python
data_args = DataArguments(text_max_length=512)
```

```python
diy_data_collator = DiyDataCollator(
    data_args=data_args,
    tokenizer=AutoTokenizer.from_pretrained("google-bert/bert-base-chinese"),
)
```

```python
data_collator_res = diy_data_collator([
    diy_dataset[0], diy_dataset[1], diy_dataset[2]
])
data_collator_res
```

## model

```python
model_args = ModelArguments(hf_name="google-bert/bert-base-chinese", num_label=4)
```

```python
model = SequenceClassificationModel(
    model_args=model_args,
    num_labels=4
)
```

```python
model(
    **data_collator_res
)
```
